```python
import pandas as pd
import numpy as np

livingtrash_df_18 = pd.read_csv("data/18년도_폐기물_처리현황_생활폐기물.csv")
livingtrash_df_17 = pd.read_csv("data/17년도_폐기물_처리현황_생활폐기물.csv")
livingtrash_df_16 = pd.read_csv("data/16년도_폐기물_처리현황_생활폐기물.csv")
livingtrash_df_15 = pd.read_csv("data/15년도_폐기물_처리현황_생활폐기물.csv")
livingtrash_df_14 = pd.read_csv("data/14년도_폐기물_처리현황_생활폐기물.csv")
```

## 전처리

```python
def living_trash_eda(df, year):
    df['종이'] = df['종이류'] + df['종이류.1']
    df['플라스틱'] = df['플라스틱류'] + df['플라스틱류.1']
    df['유리병'] = df['유리류'] + df['유리병류']

    df.drop(['종이류', '플라스틱류', '유리류', '종이류.1', '유리병류', '플라스틱류.1'], axis='columns', inplace=True)
    df.drop(['시도별(1)', '시도별(2)'], axis='columns', inplace=True)
    df.rename(columns={'시도별(3)':'처리방법'}, inplace = True)
    df['년도'] = year
```

```python
living_trash_eda(livingtrash_df_18, 2018)
livingtrash_df_18 = livingtrash_df_18.iloc[1:4]

living_trash_eda(livingtrash_df_17, 2017)
livingtrash_df_17 = livingtrash_df_17.iloc[1:4]

living_trash_eda(livingtrash_df_16, 2016)
livingtrash_df_16 = livingtrash_df_16.iloc[1:4]

living_trash_eda(livingtrash_df_15, 2015)
livingtrash_df_15 = livingtrash_df_15.iloc[1:4]

living_trash_eda(livingtrash_df_14, 2014)
livingtrash_df_14 = livingtrash_df_14.iloc[1:4]
```

```python
display(livingtrash_df_18)
display(livingtrash_df_17)
display(livingtrash_df_16)
display(livingtrash_df_15)
display(livingtrash_df_14)
```

```python
livingtrash_df = pd.concat([livingtrash_df_14,livingtrash_df_15,livingtrash_df_16,livingtrash_df_17,livingtrash_df_18])
#display(livingtrash_df)

paper_df = livingtrash_df.loc[:,['처리방법','종이', '년도']]
plastic_df = livingtrash_df.loc[:,['처리방법','플라스틱', '년도']]
bottle_df = livingtrash_df.loc[:,['처리방법','유리병', '년도']]

display(paper_df)
display(plastic_df)
display(bottle_df)
```

# 종이류 배출 추이

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(1, 1, 1)

sns.barplot(x='년도', y='종이', hue='처리방법', palette='Set3', data=paper_df, ax=ax1)

for p in ax1.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax1.annotate("%.1f 톤/일"%(height),(left+width/2, height*1.02), ha='center', va='center')

plt.title('생활폐기물 배출 추이(종이류)')
plt.ylabel('톤/일')
plt.show()
```

# 종이류 재활용률 변화

```python
paper_re_df = paper_df.loc[3]

paper_re_df.set_index('년도', inplace=True)
paper_re_df = paper_re_df.loc[:,['종이']]

paper_rate_df = paper_df.groupby(['년도']).sum()
# df_test = paper_df.transpose()
paper_rate_df = pd.merge(paper_rate_df, paper_re_df, on='년도')
paper_rate_df.columns = ['전체','재활용']
paper_rate_df['재활용률'] = round((paper_rate_df['재활용']/paper_rate_df['전체'])*100,2)
display(paper_rate_df)
```

# 플라스틱류 배출 추이

```python
fig = plt.figure(figsize=(20, 10))
ax2 = fig.add_subplot(1, 1, 1)

sns.barplot(x='년도', y='플라스틱', hue='처리방법', palette='Set3', data=plastic_df, ax=ax2)

for p in ax2.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax2.annotate("%.1f 톤/일"%(height),(left+width/2, height*1.02), ha='center', va='center')

plt.title('생활폐기물 배출 추이(플라스틱류)')
plt.ylabel('톤/일')
plt.show()
```

# 플라스틱류 재활용률 변화

```python
plastic_re_df = plastic_df.loc[3]

plastic_re_df.set_index('년도', inplace=True)
plastic_re_df = plastic_re_df.loc[:,['플라스틱']]

plastic_rate_df = plastic_df.groupby(['년도']).sum()
plastic_rate_df = pd.merge(plastic_rate_df, plastic_re_df, on='년도')
plastic_rate_df.columns = ['전체','재활용']
plastic_rate_df['재활용률'] = round((plastic_rate_df['재활용']/paper_rate_df['전체'])*100,2)
display(plastic_rate_df)
```

# 유리병류 재활용률 변화

```python
bottle_re_df = bottle_df.loc[3]

bottle_re_df.set_index('년도', inplace=True)
bottle_re_df = bottle_re_df.loc[:,['유리병']]

bottle_rate_df = bottle_df.groupby(['년도']).sum()
bottle_rate_df = pd.merge(bottle_rate_df, bottle_re_df, on='년도')
bottle_rate_df.columns = ['전체','재활용']
bottle_rate_df['재활용률'] = round((bottle_rate_df['재활용']/bottle_rate_df['전체'])*100,2)
display(bottle_rate_df)
```

```python
recycle_rate_df = pd.concat([bottle_rate_df, paper_rate_df, plastic_rate_df])
```

### 모든 사업군의 자료가 포함된 자료여서 위의 생활 폐기물 배출과 비교할수없음

# 재활용 불가 폐기물 매립 비율

```python
def landfill_rate(df, i):
    me_df = df.loc[1]
    so_df = df.loc[2]

    me_df = me_df.loc[:,[i,'년도']]
    me_df.rename(columns={i:'매립'}, inplace = True)
    #display(me_df)

    so_df = so_df.loc[:,[i,'년도']]
    so_df.rename(columns={i:'소각'}, inplace = True)
    #display(so_df)

    result = pd.merge(me_df, so_df, on='년도')
    result['재활용불가자원 매립률'] = (result['매립'] / (result['매립'] + result['소각'])) * 100
    result['구분'] = i
    result = result.loc[:,['년도','재활용불가자원 매립률','구분']]
    
    return(result)
```

```python
paper_landfill_df = landfill_rate(paper_df, '종이')
plastic_landfill_df = landfill_rate(plastic_df, '플라스틱')
bottle_landfill_df = landfill_rate(bottle_df, '유리병')

display(paper_landfill_df)
display(plastic_landfill_df)
display(bottle_landfill_df)
```

```python
total_landfill_rate =  pd.concat([paper_landfill_df, plastic_landfill_df, bottle_landfill_df])
display(total_landfill_rate)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(10, 5))
ax5 = fig.add_subplot(1, 1, 1)

sns.lineplot(x='년도', y='재활용불가자원 매립률', hue='구분', data=total_landfill_rate, ax=ax5)
plt.xticks([2014, 2015, 2016, 2017, 2018])
plt.title('재활용 불가시 매립률')
plt.ylabel('매립률')
plt.show()
```
