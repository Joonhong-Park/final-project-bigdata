# 매립지 가용량 추이

```python
import pandas as pd
import numpy as np

landfill_df = pd.read_csv("data/매립지데이터.csv")
landfill_df2 = landfill_df.iloc[:,[1,2,3]]
landfill_df2.set_index('시점', inplace=True)
display(landfill_df2)
```

```python
from matplotlib import font_manager, rc
font_path = "../data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
landfill_df2.plot(kind='area', stacked=False, alpha=0.2, figsize=(20, 10))

plt.show()
```

# 이상치 원인 파악

```python
relocation_df = pd.read_csv("data/매립지이전.csv")
display(relocation_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 1, 1)

# 그래프 그리기
sns.lineplot(x='시점' , y="총매립 용량 (m³)", hue='시도(1)',data=relocation_df,ax=ax1)

# 출력 
plt.show()
```

## 서울 -> 인천으로 매립지 이전이 원인으로 보임

```python
landfill_cost_df = landfill_df.iloc[:,[1,4]]
display(landfill_cost_df)

import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(15, 5))
ax2 = fig.add_subplot(1, 1, 1)

# 그래프 그리기
sns.lineplot(x='시점' , y="연간유지관리비 (백만원)", data=landfill_cost_df, color='green',ax=ax2)

# 출력 
plt.show()
```
